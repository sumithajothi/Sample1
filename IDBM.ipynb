```python
# Plotting and visualization
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Top_imdb.csv')

df.plot()

plt.show()
```

```python
# To Get Missing Value in percantage
percentage_missing=df.isnull().sum()*100/len(df)
print(percentage_missing)
```

```python
# Drop all the missing Values
df.dropna(axis=0)
```

```python
# Remove Duplicate Values 
duplicate_data=df.duplicated().any()
print(duplicate_data)
```

```python
# Check the column names 
df.columns
```

```python
# df['Runtime']>=180 ['Series_Title']
```

```python
# find the Average of vote in an year
df.groupby('Released_Year')['No_of_Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Released_Year',y='No_of_Votes',data=df)
plt.title("Average Vote Displayed Year Wise")
plt.show()
```

```python
# Highest Average Revenue 
df.groupby('Released_Year')['Meta_score'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Released_Year',y='Meta_score',data=df)
plt.title("Average Gross Per Year")
plt.show()
```

```python
# find the Average of Ratings of each Director
df.groupby('Director')['IMDB_Rating'].mean().sort_values()
```

```python
# Display Top 10 Movies with their Run Time
df.columns
```

```python
top=df.nlargest(10,['Runtime']),[['Series_Title','Runtime']]
print(top)
```

```python
# # visualize the relationship between IMDB rating and Runtime
boxplot=df.boxplot(column='Meta_score', by='IMDB_Rating')
plt.title('relationship between IMDB rating and Runtime')
plt.show()
```
